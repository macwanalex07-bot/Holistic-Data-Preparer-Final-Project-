# 6. Handle missing data with:
# • Simple Imputer (numerical: mean/median).

```python
from sklearn.impute import SimpleImputer

# Mean Imputation
mean_imputer = SimpleImputer(strategy="mean")
df[['age','annual_income','credit_score']] = mean_imputer.fit_transform(
    df[['age','annual_income','credit_score']]
)

# Median Imputation (optional alternative)
median_imputer = SimpleImputer(strategy="median")
df[['loan_amount']] = median_imputer.fit_transform(df[['loan_amount']])
```

# • Simple Imputer (categorical: most frequent).

```python
cat_imputer = SimpleImputer(strategy="most_frequent")

df[['employment_type','gender']] = cat_imputer.fit_transform(
    df[['employment_type','gender']]
)
```

# • Most Frequent Category Imputation.

```python
cat_imputer = SimpleImputer(strategy="most_frequent")

df[['employment_type','gender']] = cat_imputer.fit_transform(
    df[['employment_type','gender']]
)
```

# • Missing Indicator + Random Sample Imputation.

```python
# Indicator column
df['income_missing_flag'] = df['annual_income'].isnull().astype(int)

# Random sample imputation (clean version)
missing_count = df['annual_income'].isnull().sum()

random_values = df['annual_income'].dropna().sample(missing_count, replace=True)

df.loc[df['annual_income'].isnull(), 'annual_income'] = random_values.values
```

# • KNN Imputer (multivariate).

```python
from sklearn.impute import KNNImputer

knn_imputer = KNNImputer(n_neighbors=5)

df[['annual_income','loan_amount','credit_score']] = knn_imputer.fit_transform(
    df[['annual_income','loan_amount','credit_score']]
)
```

# • MICE Algorithm.

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

mice_imputer = IterativeImputer(random_state=42)

df[['annual_income','loan_amount','credit_score']] = mice_imputer.fit_transform(
    df[['annual_income','loan_amount','credit_score']]
)
```

# Part D: Outlier Handling

```python
num_cols = ['annual_income','loan_amount','credit_score']
```

# 7. Detect and treat outliers using:
# • Z-score Method.

```python
from scipy import stats
import numpy as np

z_scores = np.abs(stats.zscore(df[num_cols]))

# Keep rows where all z-scores < 3
df_z = df[(z_scores < 3).all(axis=1)]

print("Before:", df.shape)
print("After Z-score:", df_z.shape)
```

# • IQR Method.

```python
df_iqr = df.copy()

for col in num_cols:
    Q1 = df_iqr[col].quantile(0.25)
    Q3 = df_iqr[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df_iqr = df_iqr[(df_iqr[col] >= lower) & (df_iqr[col] <= upper)]

print("After IQR:", df_iqr.shape)
```

# • Percentile Method.

```python
df_pct = df.copy()

for col in num_cols:
    lower = df_pct[col].quantile(0.01)
    upper = df_pct[col].quantile(0.99)

    df_pct = df_pct[(df_pct[col] >= lower) & (df_pct[col] <= upper)]

print("After Percentile:", df_pct.shape)
```

# • Winsorization Technique.

```python
from scipy.stats.mstats import winsorize

df_win = df.copy()

for col in num_cols:
    df_win[col] = winsorize(df_win[col], limits=[0.05, 0.05])

print("After Winsorization:", df_win.shape)
```

```python
df['join_date'] = pd.to_datetime(df['join_date'])

df['year'] = df['join_date'].dt.year
df['month'] = df['join_date'].dt.month
df['day'] = df['join_date'].dt.day
df['weekday'] = df['join_date'].dt.weekday
```

# 9. Encoding categorical variables:
# • Ordinal Encoding (education levels).

```python
from sklearn.preprocessing import OrdinalEncoder

ord_enc = OrdinalEncoder(categories=[['Primary','Secondary','Graduate','Post-Graduate']])

df['education_level'] = ord_enc.fit_transform(df[['education_level']])
```

# • Label Encoding (binary features).

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['gender'] = le.fit_transform(df['gender'])
```

```python
df = pd.get_dummies(df, columns=['region','loan_purpose'], drop_first=True)
```

```python
df['income_bin'] = pd.cut(df['annual_income'], bins=4, labels=False)
```

# • Quantile Binning.


```python
df['income_quantile_bin'] = pd.qcut(df['annual_income'], q=4, labels=False)
```

# Part G: Feature Construction & Transformation
# 12. Apply transformations:
#  FunctionTransformer → log transform, reciprocal, square root.

```python
from sklearn.preprocessing import FunctionTransformer
import numpy as np

# Log Transform
log_transform = FunctionTransformer(np.log1p)
df['log_spending'] = log_transform.fit_transform(df[['spending_ratio']])

# Square Root Transform
sqrt_transform = FunctionTransformer(np.sqrt)
df['sqrt_spending'] = sqrt_transform.fit_transform(df[['spending_ratio']])

# Reciprocal Transform (avoid division by zero)
df['reciprocal_spending'] = 1 / (df['spending_ratio'] + 1e-5)
```

# PowerTransformer → Box-Cox and Yeo-Johnson.

```python
from sklearn.preprocessing import PowerTransformer

pt = PowerTransformer(method='yeo-johnson')

df[['annual_income','loan_amount']] = pt.fit_transform(
    df[['annual_income','loan_amount']]
)
```

# Column Transformer → apply different preprocessing steps to different columns.

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder

num_cols = ['annual_income','loan_amount']
cat_cols = ['gender']

ct = ColumnTransformer([
    ('num', StandardScaler(), num_cols),
    ('cat', OneHotEncoder(), cat_cols)
])

transformed_data = ct.fit_transform(df)
```
